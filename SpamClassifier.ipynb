```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import svm
```

```python
!wget https://raw.githubusercontent.com/MelihGulum/Email-Spam-Detection/main/spam.csv
```

```python
spam = pd.read_csv('/content/spam.csv',encoding='Windows-1252',on_bad_lines="skip")
spam.head()
```

```python
spam.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'], inplace=True)
spam.rename(columns={"v1":"Label","v2":"EmailText"},inplace=True)
spam.head()
```

```python
z = spam['EmailText']
y = spam["Label"]
z_train, z_test,y_train, y_test = train_test_split(z,y,test_size = 0.2)
```

```python
cv = CountVectorizer()
features = cv.fit_transform(z_train)
```

```python
model = svm.SVC()
model.fit(features,y_train)
```

```python
features_test = cv.transform(z_test)
print("Accuracy: {}".format(model.score(features_test,y_test)))
```
